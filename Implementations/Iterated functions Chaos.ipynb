### Second Example : convergence depends on  the initial value

```python
## Example below shows no matter the initial value this function does not converge
## effect of initial state or value on convergence of dynamic system
### output depends on initalisation value
import random 
def initializer_tester_2(iteration_range):
    '''
    input : initialization value for iterated function
    output: list containing function output
    
    The aim of this function is to show that for an iterated function 
    the output depens on the initial vlaues
    
    function used is : 4 * x * (1-x) ....
    
    '''
    dynamic_system = []
    x = random.uniform(0, 1)
    print('Inital x value is:',x)
    for i in range(0,iteration_range):
        x = 4 * x * (1-x)
        dynamic_system.append(x)
         
    return dynamic_system
```

```python
# using zero as initial value is bad idea , testing the inital values
initial_values = [0,0.5,1,0.05] # initial values to be tested

zero_initial = initializer_tester_2(90)
zero_dot_five_initial = initializer_tester_2(90) 
zero_dot_seven_initial = initializer_tester_2(90)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = [int(i) for i in range(0,90)]
plt.plot(x,zero_initial,label='inital:0.575')
plt.plot(x,zero_dot_five_initial,label='inital:0.885')

plt.legend(loc='best')
plt.ylabel('ylabel')
plt.xlabel('xlabel')

plt.show()
```

```python

zero_initial = initializer_tester_3(90)
zero_dot_five_initial = initializer_tester_3(90) 
zero_dot_seven_initial = initializer_tester_3(90)
```
